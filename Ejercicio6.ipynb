```python
import numpy as np
from scipy.optimize import newton
import matplotlib.pyplot as plt

# Definición de la función
def f(x):
    return np.log(x**2 + 1) - np.exp(0.4 * x) * np.cos(np.pi * x)

# Definición de la derivada de la función
def df(x):
    return (2*x / (x**2 + 1)) - np.exp(0.4 * x) * (0.4 * np.cos(np.pi * x) - np.pi * np.sin(np.pi * x))

# Graficar la función para encontrar aproximaciones iniciales
x = np.linspace(-10, 10, 1000)
y = f(x)
plt.plot(x, y)
plt.axhline(0, color='red', linestyle='--')
plt.axvline(0, color='green', linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de f(x)')
plt.show()

# Usar Newton-Raphson para encontrar ceros
tolerance = 1e-6

# Parte a: Encontrar el único cero negativo
initial_guess_neg = -1.0
zero_neg = newton(f, initial_guess_neg, fprime=df, tol=tolerance)
print("Único cero negativo:", zero_neg)

# Parte b: Encontrar los cuatro ceros positivos más pequeños
initial_guesses_pos = [0.5, 1.5, 2.5, 3.5]
zeros_pos = [newton(f, guess, fprime=df, tol=tolerance) for guess in initial_guesses_pos]
print("Cuatro ceros positivos más pequeños:", zeros_pos)

# Parte c y d: Aproximación para el enésimo cero positivo
# Basado en la gráfica, la distancia entre ceros parece ser aproximadamente 1 unidad
# Encontrar el vigesimoquinto cero positivo

initial_guess_25th = 25.0
zero_25th = newton(f, initial_guess_25th, fprime=df, tol=tolerance)
print("Vigesimoquinto cero positivo:", zero_25th)
```
